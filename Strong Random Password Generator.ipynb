```python
def Strong_Password_Generator():
    """
    Fungsi ini akan menggenerate password gabungan dari huruf, angka, dan karakter lainnya.
    User hanya perlu memasukkan panjang password yang diinginkan. 
    File hasil password yang di generate akan disimpan dalam file dengan ekstensi .txt
    File tersebut kemudian bisa dipindahkan / disimpan ditempat yang aman :)
    """
    
    # import library random
    import random
    
    # Meminta input akun yang akan dibuat passwordnya agar lebih mudah diingat
    akun = input('Akun apa yang ingin kamu buat passwordnya? ')

    # Meminta panjagn password yang diinginkan
    panjang_pass = int(input('Panjang password yang kamu inginkan: '))    

    # variabel kondisional untuk looping sampai pengguna mendapat password yang diinginkan
    pilih = True

    # looping sampai mendapat password yang diinginkan
    while pilih:

        # Membuat variabel untuk menyimpan password
        passw = ''

        # Looping untuk menggenerate password
        for i in range(panjang_pass):
            passw += ''+chr(random.randint(33,126))

        # Menampilkan hasil password
        print('\nPassword:\n' +passw +'\n')

        # Menanyakan kepada pengguna apakah ingin menggunakan password lain atau sudah yakin dengan hasil
        # password yang baru di-generate
        yakin = int(input('Apakah kamu ingin mencoba password lain? \n1. Ya \n2. Tidak, aku ingin password ini saja' 
                          +'\nPilihan kamu: '))

        # Jika ingin mencoba generate password lain
        if yakin == 1:
            pass

        # Jika yakin dengan password yang telah di generate
        else:

            # Membuat file txt untuk menyimpan password agar tidak lupa dan mudah diingat
            f = open('password ' +akun +'.txt', 'w')

            # Mengisi file txt
            f.write('Akun\t: ' +akun 
                    +'\nPass\t: '+passw)
            
            # Menutup file txt
            f.close()   

            # Menampilkan keterangan
            print('\nPassword kamu berhasil dibuat dan disimpan!')

            # Untuk keluar dari looping
            pilih = False
```
